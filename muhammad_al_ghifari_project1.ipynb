### Checking Outliers

While exploring supermarket sales data, we examined the distribution of Gross Income and checked for potential outliers using Tukey’s Rule.
This step is crucial to ensure the reliability of further analysis (especially inferential statistics) since extreme values can distort statistical tests, affect assumptions like normality, and lead to misleading conclusions.

```python
# Select the target column
col = 'GrossIncome'

# Calculate Q1, Q3, and IQR (Interquartile Range)
Q1 = supermarket_sales[col].quantile(0.25)
Q3 = supermarket_sales[col].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries using Tukey's Rule
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers outside the Tukey bounds
outliers = supermarket_sales[(supermarket_sales[col] < lower_bound) | (supermarket_sales[col] > upper_bound)]

# Calculate the percentage of outliers
total_rows = supermarket_sales.shape[0]
outlier_percentage = (len(outliers) / total_rows) * 100

# Measure skewness and kurtosis for distribution check
gross_skew = skew(supermarket_sales[col])
gross_kurtosis = kurtosis(supermarket_sales[col])

# Display results
print(f"Outlier Analysis on '{col}':")
print(f"- Total data points   : {total_rows}")
print(f"- Outliers detected   : {len(outliers)} ({outlier_percentage:.2f}%)")
print(f"- Skewness            : {gross_skew:.2f}")
print(f"- Kurtosis            : {gross_kurtosis:.2f}")
```

<div align="justify">

**The result:**
- Out of 1,000 transactions, only 9 outliers were detected (just 0.90% of the data). This indicates that most gross income values fall within a reasonable range.
- The skewness of 0.89 suggests a slight right-skew, meaning a few high-income transactions are pulling the average upward.
- The kurtosis of -0.09 shows the distribution is close to normal, neither too peaked nor too flat.

**How does gross income vary by customer type and branch?**

To better understand the revenue dynamics, we ran a Two-Way ANOVA to test whether gross income significantly varies based on customer type (`Member` vs `Normal`), branch location (A, B, or C), or a combination of both. Before diving into the analysis, we made sure all assumptions were satisfied—including checking that the variance across groups was equal using Levene’s Test. We use Two-Way ANOVA because it allows us to simultaneously examine the main effects of two categorical variables (Customer Type and Branch) and their interaction effect on a continuous outcome variable `GrossIncome`. This method is efficient and statistically powerful for comparing group means across multiple factors.

```python
from scipy.stats import levene

# Descriptive Statistics
# Calculate mean and standard deviation for each CustomerType and Branch
desc_stats = supermarket_sales.groupby(['CustomerType', 'Branch'])['GrossIncome'].agg(['mean', 'std', 'count']).reset_index()
print("Descriptive Statistics:\n", desc_stats)

# Create boxplot to compare distribution across CustomerType and Branch
plt.figure(figsize=(10, 6))
sns.boxplot(x='Branch', y='GrossIncome', hue='CustomerType', data=supermarket_sales)
plt.title('Gross Income by Customer Type and Branch')
plt.xlabel('Branch')
plt.ylabel('Gross Income')
plt.legend(title='Customer Type')
plt.tight_layout()
plt.show()

# Levene’s Test (Homogeneity of Variance)
# Split GrossIncome by CustomerType and Branch combination
groups = [
    group['GrossIncome'].values
    for _, group in supermarket_sales.groupby(['CustomerType', 'Branch'])
]

# Perform Levene’s test
stat, p = levene(*groups)
print(f"\nLevene’s Test Result:")
print(f"Statistic = {stat:.4f}, p-value = {p:.4f}")

if p > 0.05:
    print("Variance is equal across groups (homogeneous).")
else:
    print("Variance is not equal (heterogeneous). Consider using Welch ANOVA.")
```

<div align="justify">
Based on Levene’s Test, the p-value is 0.3799, which is greater than the standard significance level of 0.05. This means that the assumption of equal variances across groups (homogeneity of variance) is met. Since this assumption holds, it is appropriate to proceed with a Two-Way ANOVA analysis in otder to determine whether customer type (Member vs. Normal) and branch (A, B or C) have a significant impact on Gross Income. This test also helps us understand if there's an interaction effect between customer type and branch. By continuing with this method, we ensure that the conclusions drawn about group differences are both valid and statistically reliable.

<div align="justify">

**Hypothesis (Two-Way ANOVA 5%)**

1. Main Effect of Customer Type 
- `H₀ : There is no difference in average gross income between Member and Normal customers.`
- `H₁ : There is a difference in average gross income between Member and Normal customers.`

2. Main Effect of Branch
- `H₀: There is no difference in average gross income across the three branches (A, B, C).`
- `H₁: At least one branch has a significantly different average gross income.`

3. Interaction Effect (Customer Type × Branch)
- `H₀: There is no interaction effect between customer type and branch on gross income, the difference between Member and Normal is the same across all branches.`
- `H₁: There is an interaction, meaning the effect of customer type depends on the branch.`

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the two-way ANOVA model
model = ols('GrossIncome ~ C(CustomerType) + C(Branch) + C(CustomerType):C(Branch)', data=supermarket_sales).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display the ANOVA results
print(anova_table)
```

Here are the key findings based on the ANOVA output:
| **Factor**                 | **F-value** | **p-value** | **Interpretation** | **Reason**                                                                                                                            |
| -------------------------- | ----------- | ----------- | ------------------ | ------------------------------------------------------------------------------------------------------------------------------------- |
| **Customer Type**          | 0.353       | 0.553       | Not significant  | The p-value > 0.05        |
| **Branch**                 | 0.866       | 0.421       | Not significant  |The p-value > 0.05      |
| **Customer Type × Branch** | 0.121       | 0.886       | Not significant  | The p-value > 0.05   |


<div align="justify">
Based on the results of the Two-Way ANOVA test, we found that there is no statistically significant difference in gross income when comparing customers by type (Member vs. Normal), by branch (A, B, or C), or even when looking at the interaction between customer type and branch.

<div align="justify">

1. For Customer Type, the p-value was 0.553, which is greater than the standard significance level of 0.05. This means we fail to reject the null hypothesis, there's no meaningful difference in gross income between Members and Normal shoppers. Both contribute relatively equally to revenue.

2. For Branch, the p-value was 0.421. Again, we accept the null hypothesis, suggesting that gross income is fairly consistent across branches, with no single branch significantly outperforming the others.

3. When examining the interaction between Customer Type and Branch, the p-value soared to 0.886, the highest of all. This strongly indicates that we accept the null hypothesis once more, customer spending behavior does not differ depending on which branch they shop at.

<div align="justify">
Overall, since all p-values were well above 0.05, we conclude that any differences observed in gross income across customer types or branches are not statistically significant, and likely occurred by chance rather than due to true patterns in behavior.

**Do Payment Methods Really Matter?**

To understand how payment methods interact with both sales distribution across branches and the actual income earned, we took a two-step inferential approach: a Chi-Square Test and a One-Way ANOVA.

**First Step**

To explore how customers prefer to pay and whether those preferences affect sales, we started by examining the relationship between branch locations and payment methods. We ran a Chi-Square test to see if payment method choices differ across Branch A, B, or C. The Chi-Square test is well-suited for this analysis because it is specifically designed to assess the relationship between two categorical variables.
- Branch is a category (A, B, or C)
- Payment Method is also a category (Cash, Credit Card, E-wallet)

**Hypothesis (Chis-Square 5%)**

- `H₀ : There is no association between branch and payment method.`
- `H₁ : There is an association between branch and payment method.`

```python
from scipy.stats import chi2_contingency, f_oneway

#Chi-Square Test (Branch vs Payment Method)

# Create contingency table
contingency_table = pd.crosstab(supermarket_sales['Branch'], supermarket_sales['Payment'])

# Run Chi-Square test
chi2, p_chi, dof, expected = chi2_contingency(contingency_table)

# Display Chi-Square result
print("Chi-Square Test (Branch vs Payment Method):")
print(f"- Chi2 Statistic      : {chi2:.4f}")
print(f"- Degrees of Freedom  : {dof}")
print(f"- P-value             : {p_chi:.4f}")
if p_chi < 0.05:
    print("There is a significant relationship between branch and payment method.\n")
else:
    print("There is no significant relationship between branch and payment method.\n")
```

<div align="justify">

Chi-Square Test Result:
- Chi2 Statistic: 3.2997
- Degrees of Freedom: 4
- P-value: 0.5090

Since the p-value (0.5090) > 0.05, we fail to reject the null hypothesis.

In simple terms:
- There is no statistically significant relationship between branch and payment method.
- Customers across branches use E-wallets, Credit Cards, and Cash fairly equally, there’s no clear regional preference.

After performing a Chi-Square Test to examine whether there's a relationship between store branch and payment method, the result showed no significant association. This means that customer payment preferences appear to be independent of branch location.

<div align="justify">

**Second Step**

**But this raised a follow-up curiosity:**

We wanted to further explore how payment methods relate to gross income. Before jumping into hypothesis testing, we first looked at the descriptive statistics such as examining the average, spread, and distribution of gross income across different payment methods. This helps us understand the basic patterns and whether any noticeable differences exist at a glance.

```python
# Descriptive Stats for Gross Income by Payment 

# Group by payment method and calculate descriptive statistics
desc_stats = supermarket_sales.groupby('Payment')['GrossIncome'].describe()

# Round the values for better display
desc_stats_rounded = desc_stats.round(2)

# Print the formatted output
print("\nDescriptive Statistics of Gross Income by Payment Method:\n")
print("{:<13} {:>7} {:>8} {:>8} {:>8} {:>8} {:>8} {:>8} {:>8}".format(
    "Payment", "Count", "Mean", "Std", "Min", "25%", "50%", "75%", "Max"
))

for payment in desc_stats_rounded.index:
    row = desc_stats_rounded.loc[payment]
    print("{:<13} {:>7} {:>8} {:>8} {:>8} {:>8} {:>8} {:>8} {:>8}".format(
        payment,
        int(row['count']),
        row['mean'],
        row['std'],
        row['min'],
        row['25%'],
        row['50%'],
        row['75%'],
        row['max']
    ))

# Visualize the distribution
sns.boxplot(x='Payment', y='GrossIncome', data=supermarket_sales)
plt.title('Gross Income by Payment Method')
plt.ylabel('Gross Income')
plt.xlabel('Payment Method')
plt.show()
```

<div align="justify">

**Do Payment Methods Impact Gross Income?**

After checking the descriptive statistics of gross income for each payment method, we moved forward with an One-Way ANOVA to determine if there are any **significant differences in gross income** between **Cash**, **Credit Card**, and **E-wallet** users.Since this involves comparing a categorical variable (Payment Method) with a numerical one (Gross Income), we then decided to use an One-Way ANOVA. This test helps determine whether the average gross income differs significantly across various payment methods (Cash, Credit Card, or E-wallet) and whether any observed differences are statistically meaningful.
</div>

**Hypothesis (One-Way ANOVA 5%)**

- `H₀ : The average gross income is the same for all payment methods.`
- `H₁ : At least one payment method has a significantly different average gross income.`

```python
# One-Way ANOVA Test (Payment Method vs Gross Income)

# Fit the model using statsmodels
anova_model = ols('GrossIncome ~ C(Payment)', data=supermarket_sales).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)

# Display ANOVA results
print("\nOne-Way ANOVA Test (Payment Method vs Gross Income):")
print(anova_table)

# Interpret result
p_anova = anova_table['PR(>F)'].iloc[0]
f_stat = anova_table['F'].iloc[0]
print(f"\n- F-statistic: {f_stat:.4f}")
print(f"- P-value    : {p_anova:.4f}")

if p_anova < 0.05:
    print("There is a significant difference in gross income among payment methods.")
else:
    print("There is no significant difference in gross income among payment methods.")
```

<div align="justify">

**ANOVA Results**
- **F-statistic**: 0.0811
- **P-value**: 0.9221

These results clearly indicate that there is **no statistically significant difference** in average gross income across different payment methods (P-value > 0.05). In simpler terms, **how a customer pays (cash, card, or e-wallet) does not influence the amount of gross income generated**. Although customers use various payment methods, their spending behavior (in terms of gross income per transaction) remains fairly consistent. The choice of payment is more likely driven by personal convenience rather than any economic factor affecting sales performance.

**Does Time of Day Impact What People Buy and Spend?**

To understand how time of day impacts customer behavior, we conducted an ANOVA test on two key metrics:
- Basket Size (measured by Quantity)
- Gross Income per transaction

Each was analyzed against the categorical variable TimeSegment (morning, afternoon, evening).

**Hypothesis (One-Way ANOVA 5%)**

- `H₀ : There is no significant difference in the average basket size (quantity) across different time segments (morning, afternoon, evening).`
- `H₁ : There is a significant difference in the average basket size across different time segments.`

```python
# --- Basket Size (Quantity) ---
model_quantity = ols('Quantity ~ C(TimeSegment)', data=supermarket_sales).fit()
anova_quantity = sm.stats.anova_lm(model_quantity, typ=2)

# --- Gross Income ---
model_income = ols('GrossIncome ~ C(TimeSegment)', data=supermarket_sales).fit()
anova_income = sm.stats.anova_lm(model_income, typ=2)

# --- Display and interpret results ---
print("ANOVA - Basket Size (Quantity):\n")
print(anova_quantity, "\n")

# Interpret Quantity ANOVA
p_qty = anova_quantity['PR(>F)'].iloc[0]
if p_qty < 0.05:
    print(f"There is a significant difference in basket size across time segments (p = {p_qty:.4f})\n")
else:
    print(f"No significant difference in basket size across time segments (p = {p_qty:.4f})\n")

print("ANOVA - Gross Income:\n")
print(anova_income, "\n")

# Interpret Gross Income ANOVA
p_income = anova_income['PR(>F)'].iloc[0]
if p_income < 0.05:
    print(f"There is a significant difference in gross income across time segments (p = {p_income:.4f})")
else:
    print(f"No significant difference in gross income across time segments (p = {p_income:.4f})")
```

<div align="justify">

**ANOVA Results**

1. Basket Size (Quantity)
The ANOVA test showed a p-value of > 0.05, indicating a statistically significant difference in basket size across different times of day. This means that customers tend to purchase different quantities depending on whether they shop in the morning, afternoon, or evening. Hypothesis decision: There is no significant difference in basket size across time segments.

2. Gross Income
The ANOVA for gross income also returned a p-value of > 0.05, which means that the average gross income per transaction varies significantly by time of day. This suggests that not only do people buy different amounts, but the value of their purchases also changes depending on the time. Hypothesis decision: There is no significant difference in gross income across time segments.

**Do Low-Rated Product Lines Affect Sales Performance?**

To understand whether customer satisfaction (measured through ratings) has an impact on sales performance, we began by exploring two key metrics at the product line level:

1. Average Rating — to reflect customer perception of product quality
2. Total Gross Income — as a measure of product line performance in terms of revenue

Analyzing data at the product line level allows us to compare the performance of different categories side by side. This approach helps identify which product lines are underperforming, both in terms of customer satisfaction and sales performance. By understanding these variations, we can make more targeted, data-driven decisions, such as where to improve product quality, repackage offerings, or launch promotions.

```python
# Group by Product Line and calculate the average rating
avg_rating = supermarket_sales.groupby('ProductLine')['Rating'].mean().sort_values()

print("Average Rating by Product Line:\n")
print(avg_rating)

# Group by Product Line and calculate the total Gross Income
total_income = supermarket_sales.groupby('ProductLine')['GrossIncome'].sum().sort_values(ascending=True)

print("\nTotal Gross Income by Product Line:\n")
print(total_income)
```

<div align="justify">

Using the average rating per product line is essential because it summarizes overall customer sentiment. It enables us to rank product lines from the best-rated to the lowest-rated, giving us a clear picture of which categories are well-received and which ones need attention. Similarly, calculating the total gross income per product line gives insight into each category’s revenue contribution. This helps determine whether higher-rated products actually lead to higher sales, a key question in linking customer satisfaction to business performance.

<div align="justify">

**Correlation Analysis**

To assess the relationship between customer ratings and sales performance, we used the Pearson correlation coefficient. This method is appropriate because both avg_rating and total_income are numerical and continuous variables. Our goal is to measure whether there is a linear relationship between these two metrics.

Additionally, the dataset is already aggregated by product line, and prior checks (including skewness and outlier analysis) suggest that the data distribution is reasonably close to normal, a key assumption for Pearson correlation. This makes Pearson not only valid but the most suitable method to quantify the strength and direction of the relationship.

<div align="justify">

**Hypothesis (Pearson 5%)**

- `H₀ : There is no significant relationship between average customer rating and total gross income across product lines.`
- `H₁ : There is a significant relationship between average customer rating and total gross income across product lines.`

```python
from scipy.stats import pearsonr

# Combine average rating and total income into one DataFrame
combined = pd.concat([avg_rating, total_income], axis=1)
combined.columns = ['avg_rating', 'total_income']

# Calculate Pearson correlation
corr, pval = pearsonr(combined['avg_rating'], combined['total_income'])

# Print result with interpretation
print(f"Pearson Correlation Coefficient : {corr:.4f}")
print(f"P-Value                         : {pval:.4f}")

# Interpret the result
if pval < 0.05:
    print("There is a statistically significant relationship between average rating and total income.")
else:
    print("The relationship between average rating and total income is not statistically significant.")
```

<div align="justify">

**Pearson Results**

1. Very Weak Correlation
The correlation coefficient of 0.0925 indicates a very weak positive relationship between average rating and gross income — close to zero. This means that higher customer ratings do not appear to be strongly associated with higher sales.

2. Statistically Not Significant,
The p-value of 0.8617 is much greater than 0.05, meaning the result is not statistically significant.
In simple terms: any small correlation observed is likely due to random chance, not a meaningful pattern in the data.
